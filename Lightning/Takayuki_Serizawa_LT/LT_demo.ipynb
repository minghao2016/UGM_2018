```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem.Draw import IPythonConsole
from rdkit.Chem import Draw
```

```python
clsdata = pd.read_csv('clustered_parse.tsv', sep='\t', header=None)
clsdata.columns=['molid', 'score', 'clsid']
chembl24 = pd.read_csv('chembl_24_1.smi', sep='\t', header=None)
chembl24.columns=['smiles', 'molid']
```

```python
#cls135 = clsdata[clsdata.clsid=='CLS_ID_135']
chembldict=dict(zip(chembl24.molid, chembl24.smiles))
```

```python
def showmols(cluster_id, num_mol=20):
    select_clst = clsdata[clsdata.clsid==cluster_id]
    ids = list(select_clst.molid)
    smilist = []
    for molid in ids:
        smi = chembldict[molid]
        smilist.append(smi)
    mollist = [Chem.MolFromSmiles(smi) for smi in smilist]
    im = Draw.MolsToGridImage(mollist[:num_mol], molsPerRow=5)
    im.show()
```

```python
#CLSTURE ID is CLS_ID_xxxx, xxxx=integer from 0 to 10000
showmols('CLS_ID_9900',10)
```
